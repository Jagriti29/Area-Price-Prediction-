```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
import pickle

#Loading Dataset 
data = pd.read_csv("areaprice071.csv")

# Features and target
x=data[["area"]]
y=data["price"]

#Visualization
plt.title("Area Price Predicion")
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(x,y)
plt.show()


#splitting the data
train_x, test_x, train_y, test_y=train_test_split(x,y,test_size=0.4,random_state=42)

# Creating Model 
model=LinearRegression()
model.fit(train_x,train_y)

# Model testing 
sc=(model.score(test_x,test_y))*100
print(f"The accuracy of this model is:{sc:.2f} %" ) 

#Saving the model as pkl file
with open('area_price_model.pkl', 'wb') as f:
    pickle.dump(model, f)

print("\nModel training has been completed and saved as area_price_model.pkl")
```
